```python
import numpy as np
from IPython import display
from notebook.services.config import ConfigManager
cm = ConfigManager().update('notebook', {'limit_output': 1000})
import numpy
numpy.set_printoptions(threshold=numpy.inf)
np.set_printoptions(edgeitems=400)
np.core.arrayprint._line_width = 400
```

```python
maps = np.zeros([320,425])
maps
```

---

# 1

```python
x_start = 395
y_start = 0
fill = 30
# right top triangle
for i in range(fill):
    for j in range(i,fill):
        maps[y_start+i][x_start+j] = 1
```

```python
display.display(maps[y_start:y_start+fill,x_start:x_start+fill])
```

---

# 2

```python
# middle bar
x_start = 105
x_end = 320
y_start = 75
y_end = 91
for x in range(x_start, x_end):
    for y in range(y_start,y_end):
        maps[y][x] = 1
```

```python
display.display(maps[y_start:y_end,x_start:x_end])
```

---

# 3

```python
# right bottom bar
x_start = 320
x_end = 425
y_start = 215
y_end = 320
for x in range(x_start, x_end):
    for y in range(y_start,y_end):
        maps[y][x] = 1
```

```python
display.display(maps[y_start:y_end,x_start:x_end])
```

---

# 4

```python
# middle triangle
x_middle = 153
x_start = 115
x_end = 190
x_range = x_end - x_start
y_start = 155
y_end = 220
y_range = y_end - y_start
        
for i in range(0,y_range):
    maps[y_start+i][x_middle] = 1        
    for j in range(1,(i*int(x_range/y_range))+1):
        maps[y_start+i][x_middle+j] = 1        
        maps[y_start+i][x_middle-j] = 1        
        
```

```python
display.display(maps[y_start:y_end,x_start:x_end])
```

---

# 5

```python
# left bot triangle
x_start = 0
x_end = 60
x_range = x_end - x_start
y_start = 250
y_end = 320
y_range = y_end - y_start
for i in range(x_start, x_end):
    for j in range(y_start+(i*int(y_range/x_range)),y_end):
        maps[j][i] = 1
```

```python
display.display(maps[y_start:y_end,x_start:x_end])
```
